```python
import pandas
import matplotlib.pyplot as plt
```

```python
results_TAS = pandas.read_csv('TAS_lock.csv')
results_TTAS = pandas.read_csv('TTAS_lock.csv')
results_ticket = pandas.read_csv('ticket_lock.csv')
```

### Benchmarks' comparison
Let's plot benchmarks' results for TAS lock, TTAS lock Ð¸ ticket lock:

```python
plt.figure(figsize=(10, 12))
plt.plot(results_TAS["Number of threads"], results_TAS["time"], 
         label="TAS lock")
plt.plot(results_TTAS["Number of threads"], results_TTAS["time"], 
         label="TTAS lock")
plt.plot(results_ticket["Number of threads"], results_ticket["time"], 
         label="ticket lock")
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('number of threads', fontsize=15)
plt.ylabel('time', fontsize=15)
plt.title("Comparison of spinlocks' perfomance", fontsize=20)
plt.legend(fontsize=14, loc='best')
plt.show()
```

 We see that time of TAS lock lock/unlock rises faster than time of TTAS with increase in threads' number. Also we see that ticket lock is seemingly slower than TTAS lock and even TAS with high threads' number.
