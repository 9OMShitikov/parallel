```python
import pandas
import matplotlib.pyplot as plt
```

```python
results_PushPopDivided = pandas.read_csv('PushPopDivided.csv')
results_PushPopMixed = pandas.read_csv('PushPopMixed.csv')
```

### Comparison of different scenarios
Let's plot benchmarks' results for a large number of push, and then a large number of pop and results for the same amount of push and pop operations randomly mixed:

```python
plt.figure(figsize=(10, 8))
plt.plot(results_PushPopDivided["Number of threads"], results_PushPopDivided["time"], 
         label="Push and pop divided")
plt.plot(results_PushPopMixed["Number of threads"], results_PushPopMixed["time"], 
         label="Push and pop mixed")
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('number of threads', fontsize=15)
plt.ylabel('time', fontsize=15)
plt.title("Speed of same amount of operations with increasing number of threads", fontsize=17)
plt.legend(fontsize=14, loc='best')
plt.grid()
plt.show()
```

 We see that with mixed push and pop stack work slower (probably because of increased memory synchronization).
