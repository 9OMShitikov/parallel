```python
import pandas
import matplotlib.pyplot as plt
```

```python
results = pandas.read_csv('results.csv')
```

Let's compare acceleration compared to naive version of algorithm with different number of threads.

```python
plt.figure(figsize=(10, 12))
plt.plot(results["Number of threads"], results["time ratio"], 
         label="acceleration")
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('number of threads', fontsize=15)
plt.ylabel('acceleration', fontsize=15)
plt.title("acceleration compared to naive version of algorithm", fontsize=20)
plt.legend(fontsize=14, loc='best')
plt.show()
```

 We see that because of oferhead and because of too small parts of resulting matrix related to threads(resulting in cache invalidation) the fastest number of threads(4) is less than number of CPU cores(8).
